**Decoding**

```python
cd.replace({'Fuel_Type':{'petrol':0,'Diseal':1,'CNG':2}}, inplace=True)
```

```python
cd
```

```python
cd.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}}, inplace=True)
```

```python
cd.replace({'Seller_Type':{'Dealer':0,'Individual':1,}}, inplace=True)
```

```python
cd.replace({'Transmission':{'Manual':0,'Automatic':1,}}, inplace=True)
```

```python
cd
```

```python
x=cd.drop(['Car_Name','Selling_Price'],axis=1)
y=cd['Present_Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,y_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
tdp=model.predict(x_train)
```

```python
tda=metrics.r2_score(y_train,tdp)
tda
```

```python
tdp2=model.predict(x_test)
tdaa=metrics.r2_score(y_test,tdp2)
tdaa
```
